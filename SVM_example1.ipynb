```python
import pandas as pd

dataf_svm = pd.read_csv("E:\\Monali\\Class\\machineLearning\\SVM\\svm_dataset.csv")

dataf_svm.head()
```

```python
P = dataf_svm.drop('price', axis=1)

q = dataf_svm['price']


print('Shape of P = ', P.shape)

print('Shape of q = ', q.shape)
```

```python
from sklearn.model_selection import train_test_split

P_train, P_test, q_train, q_test = train_test_split(P, q, test_size=0.2, random_state=51)


print('Shape of P_train = ', P_train.shape)

print('Shape of q_train = ', q_train.shape)

print('Shape of P_test = ', P_test.shape)

print('Shape of q_test = ', q_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(P_train)

P_train = sc.transform(P_train)

P_test = sc.transform(P_test)
```

```python
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf')

svr_rbf.fit(P_train, q_train)

svr_rbf.score(P_test, q_test)
```

```python
svr_linear = SVR(kernel='linear')

svr_linear.fit(P_train, q_train)

svr_linear.score(P_test, q_test)
```

```python
svr_poly = SVR(kernel='poly',degree=2,)

svr_poly.fit(P_train, q_train)

svr_poly.score(P_test, q_test)
```
